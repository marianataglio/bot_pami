### Read pdf
1. Convert to .txt
2. Parse .txt

```python
from pdfminer.high_level import extract_text
import os
```

```python
input_folder = "data"
output_folder = os.path.join(input_folder, "txt")

os.makedirs(output_folder, exist_ok=True)

def convert_pdfs_to_txt(input_folder, output_folder):
    for file_name in os.listdir(input_folder):
        if file_name.endswith(".pdf"):
            input_path = os.path.join(input_folder, file_name)
            output_path = os.path.join(output_folder, file_name.replace(".pdf", ".txt"))
            
            # skip if .txt file already exists
            if os.path.exists(output_path):
                print(f"File already converted, skipping: {output_path}")
                continue
            
            try:
                text = extract_text(input_path)
                with open(output_path, "w", encoding="utf-8") as txt_file:
                    txt_file.write(text)
                print(f"Converted: {file_name} -> {output_path}")
            except Exception as e:
                print(f"Error converting {file_name}: {e}")

convert_pdfs_to_txt(input_folder, output_folder)
```
