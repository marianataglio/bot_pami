### Summarize unfiltered sections from the Boletin Oficial
1. Cada archivo en el txt contiene un indice y sus respectivos cuerpos.
2. Resumir el contenido del cuerpo. Matchear titulo con cuerpo? Ignorar titulo en el indice?

```python
import re
from collections import defaultdict
import os
```

```python
input_directory = "data/unfiltered_sections"
output_directory = os.path.join(input_directory, "merged_files")

os.makedirs(output_directory, exist_ok=True)
```

```python
for filename in os.listdir(input_directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(input_directory, filename)
        with open(file_path, "r", encoding="utf-8") as f:
            content = f.read()
            
        sections = content.strip().split('--------------------------------------------------------------------------------')
        resolutions = defaultdict(list)

        for section in sections:
            # match resolution/dispositions/circular/document identifiers
            match = re.search(r'([A-Z]+-\d{4}-\d+-INSSJP(?:-[A-Z]+)?#INSSJP)', section)
            if match:
                identifier = match.group(1)
                # add to dictionary, differentiating by content type
                resolutions[identifier].append(section.strip())

        # combine all matched sections for each identifier
        combined_sections = []
        for identifier, parts in resolutions.items():
            combined_text = "\n\n".join(parts)
            combined_sections.append(combined_text)

        result = "\n\n---\n\n".join(combined_sections)
        
        output_file_path = os.path.join(output_directory, filename)
        with open(output_file_path, "w", encoding="utf-8") as out_file:
            out_file.write(result)

        print(f"Processed and saved file: {output_file_path}")
```
